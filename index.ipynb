```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
df.shape
```

```python
# Investigate the Data
display(df.head())
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['LotArea', 'SalePrice', 'Neighborhood', 'OverallCond', 'YearBuilt', 'YrSold']], figsize=(13,13));
```

```python
# Create a plot that shows the SalesPrice Distribution
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(df['SalePrice'], bins='auto')
ax.set_title('Distribution of Sale Prices')
ax.set_xlabel('Sale Price')
ax.set_ylabel('Number of Houses')
ax.set_title('Sales Price Distribution')
ax.axvline(df['SalePrice'].mean(), color='black');
ax.axvline(df['SalePrice'].median(), color='green');
```

```python
# Create a plot that shows the Distribution of the overall house condition
fig, ax = plt.subplots(figsize=(12,6))
ax.hist(df['OverallCond'], bins='auto')
ax.set_title('Distribution of Overall House Condition')
ax.set_xlabel('Overall Contition')
ax.set_ylabel('Number of Houses')
ax.axvline(df['OverallCond'].mean(), color='black')
ax.axvline(df['OverallCond'].median(), color='green');
```

```python
# Create a Box Plot for SalePrice
fig, ax = plt.subplots(figsize=(10,8))
ax.boxplot(df['SalePrice'])
ax.set_title('Sale Price')
ax.set_ylabel('House Price (in dollars)');
```
